```python
import pure_cv as vc
from moveread.boxes import Annotations, export, Rectangle
from scoresheet_models import models
import robust_extraction as re
from robust_extraction import templates as ts
```

```python
sheet = vc.imread('../.images/models/fcde/l.jpg')
vc.show(sheet)
```

```python
# BOX COORDS
res = re.extract(sheet, ts.models['fcde']).unsafe()
vc.show(vc.draw.contours(res.contours, res.corr_img))
```

```python
ann = Annotations(box_contours=res.contours)
boxes = export(res.corr_img, ann).unsafe()
vc.show(*boxes[:16], ncols=2, height=1.5)
```

```python
# MANUAL COORDS
ann = Annotations(grid_coords=Rectangle(tl=(0.05, 0.195), size=(0.935, 0.66)))
boxes = export(sheet, ann, model=models['fcde']).unsafe()
vc.show(*boxes[:16], ncols=2, height=1.5)
```
